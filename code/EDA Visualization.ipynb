```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("../datasets/train_1st_clean.csv")
df.shape
```

```python
df = df.drop(columns='Unnamed: 0')
df.head()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr().abs()[['saleprice']].sort_values('saleprice').iloc[-13:-1], annot=True, vmin=0, vmax=1, cmap='coolwarm', annot_kws={"size":15, "va": "center_baseline", "color":"black"},)
plt.title('Abs of Top 12 Correlations with Sales Price', size=15);
plt.savefig('../visualizations/Top 12 Correlations with Sales Price.png', bbox_inches='tight')
# https://stackoverflow.com/questions/62531728/change-color-according-to-conditions-for-seaborn-heatmaps
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr().abs()[['saleprice']].sort_values('saleprice').iloc[-7:-1], annot=True, vmin=0, vmax=1, cmap='coolwarm', annot_kws={"size":15, "va": "center_baseline", "color":"black"},)
plt.title('Abs of Top 6 Correlations with Sales Price', size=15)
plt.yticks(rotation=45)
plt.savefig('../visualizations/Abs of Top 6 Correlations with Sales Price.png', bbox_inches='tight')
# https://stackoverflow.com/questions/62531728/change-color-according-to-conditions-for-seaborn-heatmaps
```
