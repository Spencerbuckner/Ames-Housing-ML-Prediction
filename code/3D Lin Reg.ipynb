```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("../datasets/train_1st_clean.csv")
df.shape
```

```python
df = df.drop(columns='Unnamed: 0')
```

# Need to feature engineer 1 column with quality

```python
df.columns
```

```python
df['exter_qual'].isna().sum()
```

```python
df['overall_qual'].isna().sum()
```

```python
df['heating_qc'].isna().sum()
```

```python
df['kitchen_qual'].isna().sum()
```

# No nulls in columns to be multiplied together

# Need to remap everything but overall_qual, since they are categorical

```python
df['exter_qual'] = df['exter_qual'].map({'Ex': 10, 'Gd': 8, 'TA': 6, 'Fa': 4, 'Po': 2})
```

```python
df['heating_qc'] = df['heating_qc'].map({'Ex': 10, 'Gd': 8, 'TA': 6, 'Fa': 4, 'Po': 2})
```

```python
df['kitchen_qual'] = df['kitchen_qual'].map({'Ex': 10, 'Gd': 8, 'TA': 6, 'Fa': 4, 'Po': 2})
```

# Now to feature engineer new column

```python
df['total_quality'] = df['exter_qual'] * df['overall_qual'] * df['heating_qc'] * df['kitchen_qual']
```

```python
df['total_quality'].describe()
```

# Now for Linear Reg

```python
features = ['total_quality', 'gr_liv_area']
```

```python
X = df[features].values.reshape(-1,2)
Y = df['saleprice']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, random_state = 42)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train), lr.score(X_test, y_test)
```

```python
cross_val_score(lr, X_train, y_train).mean()
```

# Now to make a 3D model

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = X[:, 0]
y = X[:, 1]
z = Y
```

```python
df['total_quality'].describe()
```

```python
df['gr_liv_area'].describe()
```

```python
x_pred = np.linspace(0, 10_500, 30)      # range of total_qual values
y_pred = np.linspace(300, 5750, 30)  # range of gr_liv_area values
xx_pred, yy_pred = np.meshgrid(x_pred, y_pred)
model_viz = np.array([xx_pred.flatten(), yy_pred.flatten()]).T
```

```python
from sklearn import linear_model
```

```python
ols = linear_model.LinearRegression()
model = ols.fit(X, Y)
predicted = model.predict(model_viz)
```

```python
r2 = model.score(X, Y)
r2
```

```python
plt.style.use('default')

fig = plt.figure(figsize=(18, 6))

ax = plt.axes(projection='3d')

ax.plot(x, y, z, color='k', zorder=15, linestyle='none', marker='o', alpha=0.5)
ax.scatter(xx_pred.flatten(), yy_pred.flatten(), predicted, facecolor=(0,0,0,0), s=20, edgecolor='#70b3f0')
ax.set_xlabel('Total Quality', fontsize=12)
ax.set_ylabel('gr_liv_area (sqft)', fontsize=12)
ax.set_zlabel('Saleprice ($)', fontsize=12)
ax.locator_params(nbins=4, axis='x')
ax.locator_params(nbins=5, axis='x')
ax.title.set_text('Total Quality and Living Area Size Affect on Sales Price \n R^2 = .77')

ax.view_init(elev=10, azim=112)

fig.tight_layout()
```
