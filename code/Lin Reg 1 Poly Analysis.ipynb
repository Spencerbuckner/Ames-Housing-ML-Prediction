```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("../datasets/train_1st_clean.csv")
df.shape
```

```python
kaggle_data = pd.read_csv('../datasets/test.csv')
kaggle_data.shape
```

```python
df.head()
```

```python
df = df.drop(columns='Unnamed: 0')
df.head()
```

```python
kaggle_data.columns = [col.replace(" ", "_").lower() for col in kaggle_data.columns]
kaggle_data.head()
```

# Need to feature engineer the poly columns in DF

```python
features = ['totrms_abvgrd', 'mas_vnr_area', 'full_bath', 'year_remod/add', 
            'garage_yr_blt', 'year_built', '1st_flr_sf', 'total_bsmt_sf', 
            'garage_cars', 'garage_area', 'gr_liv_area', 'overall_qual']
```

```python
X = df[features]
y = df['saleprice']
```

```python
X.head()
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(include_bias=False)
poly.fit(X)
```

```python
X_poly = poly.transform(X)
X_poly
```

```python
poly.get_feature_names(features) # 90 total, may need to cut down on columns later
```

```python
pd.DataFrame(X_poly, columns=poly.get_feature_names(features))
```

```python
lr = LinearRegression()
```

```python
cross_val_score(lr, X_poly, y).mean()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, random_state = 42)
```

```python
lr.fit(X_train, y_train)
```

```python
lr.score(X_train, y_train), lr.score(X_test, y_test)
```

# Need to update Kaggle Data sheet with poly columns and NaN fixes

```python
kaggle_data['mas_vnr_area'] = kaggle_data['mas_vnr_area'].fillna(round(kaggle_data['mas_vnr_area'].mean()))
kaggle_data['garage_yr_blt'] = kaggle_data['garage_yr_blt'].fillna(kaggle_data['year_built'])
kaggle_data['total_bsmt_sf'] = kaggle_data['total_bsmt_sf'].fillna(round(kaggle_data['total_bsmt_sf'].mean()))
kaggle_data['garage_cars'] = kaggle_data['garage_cars'].fillna(round(kaggle_data['garage_cars'].mean()))
kaggle_data['garage_area'] = kaggle_data['garage_area'].fillna(round(kaggle_data['garage_area'].mean()))
```

```python
Kaggle_X = kaggle_data[features]
```

```python
Kaggle_poly = PolynomialFeatures(include_bias=False)
Kaggle_poly.fit(Kaggle_X)
```

```python
Kaggle_X_poly = Kaggle_poly.transform(Kaggle_X)
Kaggle_X_poly
```

```python
Kaggle_poly.get_feature_names(features)
```

```python
kaggle_data1 = pd.DataFrame(Kaggle_X_poly, columns=Kaggle_poly.get_feature_names(features))
```

# Need to rearrange ID columns

```python
kaggle_data1['id'] = kaggle_data['id']
kaggle_data1.head()
```

```python
cols = kaggle_data1.columns.tolist()
cols = cols[-1:] + cols[:-1]
#Inspiration https://stackoverflow.com/questions/13148429/how-to-change-the-order-of-dataframe-columns
```

```python
kaggle_data1 = kaggle_data1[cols].copy() #  .copy() gets rid of the data error problem when making the saleprice column
kaggle_data1
```

# Make predictions off train set on kaggle data

```python
preds =lr.predict(kaggle_data1[Kaggle_poly.get_feature_names(features)])
```

```python
kaggle_data1['saleprice'] = preds
```

```python
kaggle_data1
```

```python
final_kaggle =kaggle_data1[['id', 'saleprice']]
```

```python
final_kaggle.to_csv('../datasets/kaggle_mlr1_poly.csv', index=False)
```
